#### Definimos el sistema de ecuaciones con las modificaciones adecuadas

```python
def F(X, t, p, N, nICU, switchLKDWN, nWeek):
    S, E, I, L, G, H, ICU, R, M = X
        
    if ICU > nICU/(N*1e6):
        P_M = p['P_M']*2
    else:
        P_M = p['P_M']
    
    if switchLKDWN:
        week = nWeek*7
        if (t % (2*week)) > week: # (((t % (2*week)) > week) & (t <= 205)): #t > 80 and t < 120:
            R0 = p['R0_2']
        else:
            R0 = p['R0_1']
    
    elif type(nWeek) == list:
        if t > nWeek[0] and t < nWeek[1]:
            R0 = p['R0_2']
        else:
            R0 = p['R0_1']
    else:
         R0 = p['R0_1']

    dS = -(R0/p['D_inf'])*I*S
    dE = (R0/p['D_inf'])*I*S - (1/p['D_inc'])*E
    dI = (1/p['D_inc'])*E - (1/p['D_inf'])*I
    dL = (1-p['P_grave'])*(1/p['D_inf'])*I - (1/p['D_RL'])*L
    dG = p['P_grave']*(1/p['D_inf'])*I - (1/p['D_hosp'])*G
    dH = (1/p['D_hosp'])*G - (1-p['P_ICU'])*(1/p['D_RH'])*H - p['P_ICU']*(1/p['D_ICU'])*H
    dICU = p['P_ICU']*(1/p['D_ICU'])*H - (1-P_M)*(1/p['D_RICU'])*ICU - P_M*(1/p['D_M'])*ICU
    dR = (1/p['D_RL'])*L + (1-p['P_ICU'])*(1/p['D_RH'])*H + (1-P_M)*(1/p['D_RICU'])*ICU
    dM = P_M*(1/p['D_M'])*ICU

    return [dS, dE, dI, dL, dG, dH, dICU, dR, dM]
```

#### Generamos un objeto que permitira instanciar diversos modelos con sus metodos y funciones para un manejo mas sencillo

```python
class epDynamic():
    ''' 
        Variables/Estados:
            N = 22 (Poblacion susceptible en millones de acuerdo a la CONAPO (2019))
            S = Susceptibles
            E = Expuestos
            I = Infectados
            L = Contagiados sintomaticos leves
            G = Contagiados sintomaticos graves
            H = Hospitalizados
            ICU = Unidad de terapia intensiva
            R = Recuperados
            M = Muertos
        Parametros del modelo:
            D_inf = Tiempo que es infeccioso el paciente 
            D_inc = Tiempo de incubacion (dias) 
            P_grave = Tasa de hospitalizacion
            D_RL = Tiempo de recuperacion de casos leves (dias) 
            D_hosp = Tiempo entre presencia de sintomas y hospitalizacion en casos graves (dias)
            P_ICU = Tasa de UCI 
            D_RH = Tiempo entre hospitalizacion de casos graves no UCI y recuperacion (dias)
            D_ICU = Tiempo entre hospitalizacion de casos graves e ingreso a UCI (dias) 
            P_M = Tasa de letalidad respecto a poblacion general (0.3 reportado, 0.9 estimado en Mexico)
            D_RICU = Tiempo entre ingreso a UCI y recuperacion (dias) 
            D_M = Tiempo entre ingreso a UCI y deceso (dias) 
            R0_1 = 2.83 Inferido, pero consistente con Leung M. & Leung G. (2020)
            R0_2 = Valor R0 esperado durante el periodo de cuarentena
        Otros parametros:
            T = Tiempo de integracion (dias)
            nT = Steps de integracion
            nICU = Numero de camas ICU
            switchLKDWN = (Bool) Si la cuarentena sera periodica (True) o si sera un evento unico (False)
            nWeek = Si switchLKDWN es True: Se asigna un numero que corresponde a la periodicidad de la cuarentena, int
                            E.g: 1 = Una semana sin cuarentena, una semana con cuarentena ... Una semana sin cuarentena, una semana con cuarentena
                                 3 = Tres semanas sin cuarentena, Tres semana con cuarentena ... Tres semanas sin cuarentena, Tres semanas con cuarentena
                    Si switchLKDWN es False: Un vector o tupla con los dias de inicio y fin de la cuarentena [inicio,fin]
                        
        '''
    def __init__(self, N = 22, E = 0, I = 600*1e-6, L = 0, G = 0, H = 0, ICU = 0, R = 0, M = 0, R0_1 = 2.83, R0_2 = 0.95, 
                 D_inf = 2.9, D_inc = 5.2, P_grave = 0.138, D_RL = 14, D_hosp = 4, P_ICU = 0.05, D_RH = 12, D_ICU = 1, 
                 P_M = 0.09, D_RICU = 7, D_M = 8, T = 365., nT= 1200, nICU = 22500, switchLKDWN = False, nWeek = False, pEconomic = 0.3,
                title = False, xlab = False, ylab = False, legend = False):
        self.t = np.linspace(0, T, nT)
        self.x0 = [(N-I)/N, E/N, I/N, L/N, G/N, H/N, ICU/N, R/N, M/N]
        self.p = {'R0_1': R0_1, 'R0_2': R0_2,'D_inf': D_inf, 'D_inc': D_inc, 'P_grave': P_grave, 'D_RL': D_RL, 'D_hosp': D_hosp, 'P_ICU': P_ICU, 'D_RH': D_RH,'D_ICU': D_ICU, 'P_M': P_M, 'D_RICU': D_RICU, 'D_M': D_M}
        self.xt = odeint(F, self.x0, self.t, args = (self.p, N, nICU, switchLKDWN, nWeek))
        self.x_ind = ['Susceptibles', 'Exposed', 'Infected', 'Mild Symp.', 'Severe Symp.', 'Hospitalized', 'ICU', 'Recovered', 'Deceased']
        self.susceptibles = self.xt[:,0]
        self.exposed = self.xt[:,1]
        self.infected = self.xt[:,2]
        self.mild_symp = self.xt[:,3]
        self.severe_symp = self.xt[:,4]
        self.hospitalized = self.xt[:,5]
        self.ICU = self.xt[:,6]
        self.recovered = self.xt[:,7]
        self.deaths = self.xt[:,8]
        
        self.switchLKDWN = switchLKDWN
        self.nWeek = nWeek
        
        self.ecoActive = self.susceptibles + self.exposed + self.recovered
        # self.ecoActiveWMean = np.append(self.ecoActive, self.ecoActive.mean())
        self.ecoInactive = self.infected + self.mild_symp + self.severe_symp + self.hospitalized + self.ICU + self.deaths
        
        if self.switchLKDWN:
            week = self.nWeek*7
            ## Identificar el dia en el que la solucion es estable, no solo usar 205, tambien debajo en la grafica
            self.propEconomy = [active*proportion for active, proportion in zip (self.ecoActive, [pEconomic if i == True else 1 for i in (((self.t % (2*week)) > week) & (self.t <= 205))])]
        elif type(self.nWeek) == list:
            self.propEconomy = [active*proportion for active, proportion in zip(self.ecoActive, [pEconomic if i == True else 1 for i in ((self.t >= self.nWeek[0]) & (self.t <= self.nWeek[1]))])]
        else:
            self.propEconomy = self.ecoActive
        
        ##
        self.title = title
        self.xlab = xlab
        self.ylab = ylab
        self.legend = legend
        
    def plotModel(self): #, self.xt, self.switchLKDWN, self.nWeek):  
        if self.xlab:
            plt.xlabel('Time (days)')
        if self.ylab:
            plt.ylabel('Proportion of the population')

        for i in range(len(self.x0)):
            plt.plot(self.t, self.xt[:,i], label = self.x_ind[i])

        plt.grid(False)#(True)
        if self.switchLKDWN:
            if self.title:
                plt.title('Lockdown every {} weeks'.format(self.nWeek))
            week = self.nWeek*7
            plt.fill_between(self.t, 0, 1, where= (((self.t % (2*week)) > week) & (self.t <= 205)), facecolor='green', alpha=0.2, label='Lockdown')
        elif type(self.nWeek) == list:
            if self.title:
                plt.title('Lockdown from day {} until day {}'.format(self.nWeek[0], self.nWeek[1]))
            plt.fill_between(self.t, 0,1, where = ((self.t >= self.nWeek[0]) & (self.t <= self.nWeek[1])), facecolor='green', alpha=0.2)
        else:
            if self.title:
                plt.title('Dynamic without intervention')
        if self.legend:
            plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")  # (loc = 'best')
```

#### Generamos distintas intervenciones con diferentes inicios y duraciones de la cuarentena

```python
noSwitchLK = []
for i in np.linspace(1, 63, 63):
    inicio = []
    for j in np.linspace(i+1, i+180, 180):
        inicio.append(epDynamic(nWeek = [i,j], R0_2 = 0.95,  T = 1200, nT=2400).xt)
    noSwitchLK.append(inicio)
```

```python
noSwitchLK = np.array(noSwitchLK)
```

Recuperamos los casos acumulados de ICU para cada intervencion

```python
mICUs_cumsum = np.zeros((63,180))
for i in range(63):
    for j in range(180):
        mICUs_cumsum[i][j] = np.cumsum(noSwitchLK[i][j][:, 6])[-1]
```

Visualizamos los datos:

```python
#with plt.xkcd(): #Si se utilizará la funcion anterior, todo el codigo posterior de este bloque debera estar identado a esta linea
sns.heatmap(mICUs_cumsum.T, cmap = 'plasma')
plt.yticks(np.arange(0, 190, 10), np.arange(0, 190, 10))
plt.xticks(np.arange(0, 63, 5), np.arange(0,63,5))
plt.gca().invert_yaxis()
plt.title('Cumulative ICU cases')
plt.xlabel('Days post-infection')
plt.ylabel('Duration of lockdown')
# plt.savefig('./inicioVSduracion_ICUs_cumsum.png')
plt.show()
```

#### Ahora instanciamos diversas cuarentenas utilizando la clase 'epDynamic'

##### Ahora, se accedera al vector de las personas economicamente activas y de las muertes mediante los métodos de cada cuarentena que se instanció anteriormente
